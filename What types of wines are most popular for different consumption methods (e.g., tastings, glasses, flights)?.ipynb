```python
Pearmund_Cellars = df[df['Winery'] == 'Pearmund Cellars']
Pearmund_Cellars
```

```python

df_grouped_1 = Pearmund_Cellars.groupby('Type')['Quantity'].sum().reset_index()
df_grouped_1
```

```python
wine_flight = df_grouped_1.loc[df_grouped_1['Type'].isin(['WINE FLIGHT', 'WINE FLIGHT Reserve']), 'Quantity'].sum()
new_row = pd.DataFrame({'Type': ['Wine Flight'], 'Quantity': [wine_flight]})
df_grouped_1 = pd.concat([df_grouped_1, new_row], ignore_index=True)
df_grouped_1
```

```python
index_of_wine_flight = df_grouped_1[df_grouped_1['Type'] == 'Wine Flight'].index[0]
df_grouped_1.loc[index_of_wine_flight, 'Quantity'] //= 3
```

```python
df_grouped_1
```
